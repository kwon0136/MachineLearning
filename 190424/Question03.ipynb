```
# Diabets.csv데이터를 이용하여 모델을 정의하고 예측하고, Tree그래프를 출력하시오.(Decision Tree)

# library import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```
# dataset import

dataset = pd.read_csv("Diabetes.csv")
```

```
# data, target 설정

sel_names = ['Pregnancies', 'Insulin', 'BMI', 'Age', 'Glucose', 'BloodPressure']
x = dataset[sel_names]
y = dataset["Outcome"]
```

```
# train, test data split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```
# DecisionTree class 생성

DT = DecisionTreeClassifier()
```

```
# trainning

DT.fit(xtrain, ytrain)
```

```
# xtest 값 예측

rResult = DT.predict(xtest)
rResult
```

```
# 그래프 출력

from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
import graphviz

dot_data = StringIO()
import os
os.environ['PATH'] += os.pathsep + 'C:\\Program Files (x86)\\Graphviz2.38\\bin'

```

```
export_graphviz(DT, out_file=dot_data, filled=True, rounded=True, special_characters=True,
                feature_names=sel_names, class_names=['0', '1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png())

```

```
# Car.csv데이터를 이용하여 모델을 정의하고 예측하고, Tree그래프를 출력하시오.(Decision Tree)

# library import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import preprocessing
```

```
# dataset import

dataset = pd.read_csv("Car.csv")
```

```
# data 전처리(string --> int)

le = preprocessing.LabelEncoder()

dataset['buying'] = le.fit_transform(dataset.buying)
dataset['maint'] = le.fit_transform(dataset.maint)
dataset['lug_boot'] = le.fit_transform(dataset.lug_boot)
dataset['safety'] = le.fit_transform(dataset.safety)
dataset['doors'] = le.fit_transform(dataset.doors)
dataset['persons'] = le.fit_transform(dataset.persons)
```

```
# data, target 설정

sel_names = ['buying',	'maint',	'doors',	'persons',	'lug_boot',	'safety']
x = dataset[sel_names]
y = dataset["class"]
```

```
# train, test data split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```
# DecisionTree class 생성

DT2 = DecisionTreeClassifier()
```

```
# trainning 

DT2.fit(xtrain, ytrain)
```

```
# xtest 값 예측

rResult = DT2.predict(xtest)
rResult
```

```
# 그래프 출력

from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
import graphviz

dot_data = StringIO()
import os
os.environ['PATH'] += os.pathsep + 'C:\\Program Files (x86)\\Graphviz2.38\\bin'
```

```
export_graphviz(DT2, out_file=dot_data, filled=True, rounded=True, special_characters=True,
                feature_names=sel_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png())
```
